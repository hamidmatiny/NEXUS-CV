# NEXUS-CV Trajectory LSTM Training

This notebook demonstrates training the TrajectoryLSTM model on synthetic track histories.
For production training, use `models/train_trajectory_lstm.py` with MLflow autologging.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path('..').resolve()))

import torch
from intelligence.trajectory_lstm import TrajectoryLSTM, HORIZON
from models.train_trajectory_lstm import generate_synthetic_dataset
```

```python
inputs, targets = generate_synthetic_dataset(num_sequences=500)
print(f'Inputs: {inputs.shape}, Targets: {targets.shape}')
```

```python
model = TrajectoryLSTM(horizon=HORIZON)
x = torch.tensor(inputs[:8])
with torch.no_grad():
    out = model(x)
print(f'Output shape: {out.shape}')
```
